```python
import nltk
from collections import Counter
import matplotlib.pyplot as plt
import re

# If nltk data is not downloaded yet
nltk.download('punkt')

# Sample list of sentences
sentences = [
    "This is the first sentence.",
    "And here is another sentence.",
    "This sentence is the third one."
]

# Function to tokenize sentences into words
def tokenize_sentences(sentences):
    words = []
    for sentence in sentences:
        words.extend(nltk.word_tokenize(sentence))
    return words

# Function to clean and count word frequencies
def count_word_frequencies(words):
    # Remove punctuation and convert to lowercase
    words = [re.sub(r'[^\w\s]', '', word).lower() for word in words if word.isalpha()]
    return Counter(words)

# Tokenize the sentences
words = tokenize_sentences(sentences)

# Count the word frequencies
word_freq = count_word_frequencies(words)

# Plotting the histogram
def plot_histogram(word_freq):
    plt.figure(figsize=(10, 6))
    plt.bar(word_freq.keys(), word_freq.values())
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Word Frequency Histogram')
    plt.show()

# Plot the word frequency histogram
plot_histogram(word_freq)
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Sample list of sentences and their sentiments
data = [
    ("This is the best day ever!", "positive"),
    ("I am so sad today.", "negative"),
    ("I'm feeling quite neutral about this.", "neutral"),
    ("What a wonderful experience!", "positive"),
    ("This is terrible.", "negative"),
    ("I don't know how I feel about this.", "neutral"),
    ("Absolutely fantastic!", "positive"),
    ("I'm not sure about this.", "neutral"),
    ("I hate this!", "negative"),
    ("This is okay.", "neutral")
]

# Extract the sentiments from the data
sentiments = [sentiment for _, sentiment in data]

# Count the frequency of each sentiment
sentiment_freq = Counter(sentiments)

# Plotting the bar graph
def plot_sentiment_histogram(sentiment_freq):
    plt.figure(figsize=(5, 3))
    plt.bar(sentiment_freq.keys(), sentiment_freq.values(), color=['blue', 'green', 'red'])
    plt.xlabel('Sentiment')
    plt.ylabel('Frequency')
    plt.title('Sentiment Frequency Bar Graph')
    plt.show()

# Plot the sentiment frequency histogram
plot_sentiment_histogram(sentiment_freq)
```
