```python
import pandas as pd
data1=pd.read_csv("filtered_movies")
data2=pd.read_csv("filtered_books")
```

```python
data2.head()
```

```python
import pandas as pd

# Assuming you have two dataframes: movie_df and books_df

# Merge the two dataframes on the user ID
merged_df = pd.merge(data1, data2, on='user_id', suffixes=('_movie', '_book'))

# Filter the merged dataframe based on the condition: movie timestamp < book timestamp
filtered_df = merged_df[(merged_df['timestamp_movie'] < merged_df['timestamp_book'])]

# Add the book title column to the movie dataframe as the target
data1['Target_Title'] = filtered_df['title_book']

data1.dropna(subset=['Target_Title'], inplace=True)

# Count the occurrences of each user ID
user_counts = data1['user_id'].value_counts()

# Get the user IDs where the count is less than 2
users_to_remove = user_counts[user_counts < 2].index

# Filter the dataframe to remove rows with user IDs that occur less than 2 times
filtered_data = data1[~data1['user_id'].isin(users_to_remove)]
```

```python
data1.head(10)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have loaded your datasets into pandas DataFrames: movie_df and books_df

# Sort datasets by timestamp
#data1.sort_values(by='timestamp', inplace=True)
#df2.sort_values(by='timestamp', inplace=True)

# Define the timestamp threshold for splitting (e.g., 80% for training)
timestamp_threshold = 0.8

# Function to split interactions for each user
def split_by_user(df):
    train_data = pd.DataFrame(columns=df.columns)
    test_data = pd.DataFrame(columns=df.columns)
    
    for user_id, user_data in df.groupby('user_id'):
        train, test = train_test_split(user_data, train_size=timestamp_threshold, shuffle=False)
        train_data = pd.concat([train_data, train])
        test_data = pd.concat([test_data, test])
    
    return train_data, test_data

# Split movie dataset
train, test = split_by_user(filtered_data)


# Optionally, you can reset the index if needed
#movie_train.reset_index(drop=True, inplace=True)
#movie_test.reset_index(drop=True, inplace=True)
#books_train.reset_index(drop=True, inplace=True)
#books_test.reset_index(drop=True, inplace=True)

# Now you have movie_train, movie_test, books_train, and books_test DataFrames
# containing training and testing data for movies and books datasets, respectively
```
