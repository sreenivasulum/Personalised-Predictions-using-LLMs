```python
import pandas as pd

# Assuming you have two dataframes: movie_df and books_df

# Merge the two dataframes on the user ID
merged_df = pd.merge(data1, data2, on='user_id', suffixes=('_phones', '_book'))

# Filter the merged dataframe based on the condition: movie timestamp < book timestamp
filtered_df = merged_df[(merged_df['timestamp_phones'] < merged_df['timestamp_book'])]

# Add the book title column to the movie dataframe as the target
data1['Target_Title'] = filtered_df['title_book']

data1.dropna(subset=['Target_Title'], inplace=True)

# Count the occurrences of each user ID
user_counts = data1['user_id'].value_counts()

# Get the user IDs where the count is less than 2
users_to_remove = user_counts[user_counts < 2].index

# Filter the dataframe to remove rows with user IDs that occur less than 2 times
filtered_data = data1[~data1['user_id'].isin(users_to_remove)]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming you have loaded your datasets into pandas DataFrames: movie_df and books_df

# Sort datasets by timestamp
#data1.sort_values(by='timestamp', inplace=True)
#df2.sort_values(by='timestamp', inplace=True)

# Define the timestamp threshold for splitting (e.g., 80% for training)
timestamp_threshold = 0.8

# Function to split interactions for each user
def split_by_user(df):
    train_data = pd.DataFrame(columns=df.columns)
    test_data = pd.DataFrame(columns=df.columns)
    
    for user_id, user_data in df.groupby('user_id'):
        train, test = train_test_split(user_data, train_size=timestamp_threshold, shuffle=False)
        train_data = pd.concat([train_data, train])
        test_data = pd.concat([test_data, test])
    
    return train_data, test_data

# Split movie dataset
train, test = split_by_user(filtered_data)


# Optionally, you can reset the index if needed
#movie_train.reset_index(drop=True, inplace=True)
#movie_test.reset_index(drop=True, inplace=True)
#books_train.reset_index(drop=True, inplace=True)
#books_test.reset_index(drop=True, inplace=True)

# Now you have movie_train, movie_test, books_train, and books_test DataFrames
# containing training and testing data for movies and books datasets, respectively
```

```python
import pandas as pd

def load_and_process_dataset(file_path, columns_to_keep=None, rename_columns=None):
    """
    Load a dataset from a CSV file, select specific columns, and rename them.
    
    Parameters:
        file_path (str): Path to the CSV file.
        columns_to_keep (list): List of column names to keep. If None, keep all columns.
        rename_columns (dict): Dictionary mapping old column names to new ones for renaming.
        
    Returns:
        pandas.DataFrame: Processed DataFrame.
    """
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Select specific columns if specified
    if columns_to_keep:
        df = df[columns_to_keep]
    
    # Rename columns if specified
    if rename_columns:
        df.rename(columns=rename_columns, inplace=True)
    
    return df


def merge_dataframes(left_df, right_df, on_column, how='left'):
    """
    Merge two DataFrames on a specified column.
    
    Parameters:
        left_df (pandas.DataFrame): Left DataFrame.
        right_df (pandas.DataFrame): Right DataFrame.
        on_column (str): Name of the column to merge on.
        how (str): Type of merge to be performed ('left', 'right', 'outer', 'inner').
        
    Returns:
        pandas.DataFrame: Merged DataFrame.
    """
    merged_df = pd.merge(left_df, right_df, on=on_column, how=how)
    return merged_df


# Example usage:
List1=["s41","s51","s61","s71"]
List2=[dataset4M,dataset5M,dataset6M,dataset7M]
List3=["phones","electronics","clothing","digitalmusic"]
for i,j,k in zip(List1,List2,List3):
    # Load and process first dataset
    df = load_and_process_dataset(i, columns_to_keep=["user_id", "asin", "rating", "text", "timestamp"], rename_columns={"text": "product review", "asin": "Product Id"})

    # Load and process second dataset
    #dfM = load_and_process_dataset(j, rename_columns={"parent_asin": "Product Id", "title": "Product Title"})
    # Process the secondary dataset
    dfM = pd.DataFrame(j)
    dfM = dfM[["parent_asin", "title"]]
    dfM = dfM.rename(columns={'title': 'Product Title', 'parent_asin': 'Product Id'})
    # Merge the two dataframes
    merged_df = merge_dataframes(df, dfM, on_column="Product Id", how="left")

# Display the merged DataFrame
    print(merged_df.head())
    merged_df.to_csv("{k}_products",index=False)
    
```
